# 1. Pengenalan Dataset

Dataset ini berasal dari sumber publik kaggle dan berisi informasi mengenai kualitas wine merah.
Fitur-fitur dalam dataset mencakup berbagai atribut kimia seperti kadar alkohol, keasaman, pH, sulfur dioksida, dan lainnya.
Target dari dataset ini adalah variabel 'quality', yang mencerminkan skor kualitas wine berdasarkan penilaian sensorik.
Dataset ini cocok untuk permasalahan klasifikasi atau regresi, tergantung bagaimana variabel target diolah. berikut linknya (https://www.kaggle.com/datasets/yasserh/wine-quality-dataset)


# 2. Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

# 4. Exploratory Data Analysis (EDA)


```python
# 4. Exploratory Data Analysis (EDA)
# Mengecek jumlah missing values
print("\nMissing values per kolom:")
print(df.isnull().sum())

# Distribusi target variabel
plt.figure(figsize=(8, 5))
sns.countplot(x='quality', data=df)
plt.title('Distribusi Kualitas Wine')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()

# Korelasi antar fitur
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi Antar Fitur')
plt.show()

# Boxplot untuk mendeteksi outlier
num_features = df.select_dtypes(include=[np.number]).columns.tolist()
plt.figure(figsize=(20, 15))
for i, col in enumerate(num_features):
    plt.subplot(4, 4, i+1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot {col}')
plt.tight_layout()
plt.show()
```

# 5. Data Preprocessing

```python
# 5. Data Preprocessing
# Menghapus duplikasi
df = df.drop_duplicates()

# Menangani missing values (jika ada)
df = df.dropna()  # atau gunakan imputasi jika diperlukan

# Standarisasi fitur numerik
scaler = StandardScaler()
fitur = df.drop("quality", axis=1)
fitur_scaled = scaler.fit_transform(fitur)

# Gabungkan kembali dengan target
df_processed = pd.DataFrame(fitur_scaled, columns=fitur.columns)
df_processed['quality'] = df['quality'].values

# Simpan data hasil preprocessing
import os
os.makedirs("../preprocessing/namadataset_preprocessing", exist_ok=True)
df_processed.to_csv("WineQT_processed.csv", index=False)

print("\nPreprocessing selesai. Dataset siap digunakan untuk pelatihan model.")
```
